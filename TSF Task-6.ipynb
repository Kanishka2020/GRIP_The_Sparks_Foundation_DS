###Importing Libraries

```
import sklearn.datasets as datasets
import pandas as pd
```

###Loading the iris dataset

```
#Load the dataset
iris=datasets.load_iris()

#Frame the iris dataframe
x=pd.DataFrame(iris.data, columns=iris.feature_names)
x.head()
```

```
y=iris.target
y
```

##Describe the dataset

```
x.describe()
```

##Train and Split the dataset

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

####Now we will import Decision Tree Classifier and fit the model

```
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(x_train,y_train)
```

```
y_predict = dtree.predict(x_test)
y_predict
```

##Constructing Decision Tree using Matplotlib.pyplot

```
import matplotlib.pyplot as plt
from sklearn import tree
a=['sepal length (cm)','sepal width (cm)','petal width (cm)', 'petal_width (cm)']
b=['setosa','versicolor','virginica']

fig, axes = plt.subplots(nrows = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(dtree, feature_names = a, class_names = b, filled = True)
```
